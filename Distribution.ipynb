```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

We use Distribution plots to visualize the distribution of quantitative Data.  

There are some plots that allows distribution of a data set.
- displot/histplot
- jointplot
- pairplot
- rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

- displot/histoplot

```python
plt.subplot(1,2,1,)
sns.histplot(df['total_bill'], bins=10, kde= True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], kde= True)
```

- jointplot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'scatter')
```

- pairplot

```python
sns.pairplot(df, hue= 'sex', palette='rainbow')
```

- rugplot

```python
sns.rugplot(df['total_bill'])
```
