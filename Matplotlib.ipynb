# Matplotlib Basics

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Learning")
plt.xlabel("Starting")
plt.ylabel("growth")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2) # here we have to plot values.
plt.subplot(2,2,4)
plt.plot(x**3,y) 
```

### Object Oriented Way

```python
# Here we get controlled on canvas to design it by your prespective.
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.6, 0.6])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2, 0.35, 0.27, 0.3])
axis2.plot(x,y)
```

Matplotlib allows the aspect ratio, DPI & figure size to be sepecified when the Figure object is Created.

Figsize is a tuple of the width & height of the figure in inches DPI is the dots-per-inch (Pixel per inch) Dpi increase quality.

```python
fig = plt.figure(figsize=(6,8), dpi=150) # 6 means 600 & 8 is 800.
axis3 = fig.add_axes([0.1, 0.1, 0.7, 0.4])
axis3.plot(x,y)
```
